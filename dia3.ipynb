### üß† EJERCICIO ‚Äî Funci√≥n con l√≥gica real (nivel alto)
#### üìå Enunciado
Crea una funci√≥n llamada analizar_numeros que reciba una lista de n√∫meros y devuelva tres cosas:
- La suma total
- El promedio
- La cantidad de n√∫meros pares
‚ö†Ô∏è Reglas:
- Usa una sola funci√≥n
- Usa al menos un for
- La funci√≥n debe retornar los valores (no imprimirlos)

```python
#Creamos la funcion analizar_numeros cuyo parametros es una lista
def analizar_numeros (lista):
    suma = 0                        #Inicializamos suma (suma de los elementos de la lista) igual 0, necesitamos para sacar el promedio
        
    cant_pares = 0                  #Inicializamos el contador de pares en 0

    for i in lista:                #Recorremos la lista con la interacion i
        suma += i                   #Suma es igual a suma=0 + i la primera iteracion + el elemento en posicion i
        if i%2 == 0:                #Si i modulo 2 es 0 significa que es un numero par
            cant_pares += 1         #El contador cant_pares suma 1 unidad

    promedio = suma/len(lista)          #El promedio es la suma de los elementos de la lista/cantidad de elementos(len)
    return (suma, promedio, cant_pares) #La funcion nos devuelve los valores de suma, promedio, cantidad de pares en una lista 

#Creamos una lista de numeros
lista_num = [10, 35, 55, 20, 65, 70]

#Llamamos a la fucioin y guardamos en una lista
resultado = analizar_numeros(lista_num)

#imprimir en pantalla los resultados
print('La suma es:', resultado[0])
print('El promedio es:', resultado[1])
print('Cantidad de numeros pares:', resultado[2])
```

### Challenge 3 ü§∫ - 15m
Crear una funci√≥n que reciba dos par√°metros (una palabra y un n√∫mero), y que retorne True si el n√∫mero es igual a la cantidad de letras de la palabra. En caso contrario, que retorne False.

```python
#Pedimos al usuario que ingrese una palabra y un numero

palabra = str(input('Ingrese una palabra: '))
numero = int(input('Ingrese un numero: '))

#Definimos una funcion que verifique que la longitud de la palabra coincide con el numero
def verificar(letras, num):
    if len(letras)== num:
        return True
    else:
        return False
    
#Llamamos a la funcion
verificar(palabra, numero)
```

#### Challenge 4 ü§∫ - 20 min
Crear funci√≥n que revise si una palabra contiene una letra. En caso de que la contenga, retornar la letra entre corchetes []. De lo contrario, retornar solo la letra.

```python
#Ingresamos una palabra y la guardamos
palabra = input('Ingrese una palabra: ')
letra = input('Ingrese una letra:')

#Definimos una funcion con 2 parametros
def revisar (letras, c):       
    for i in palabra:
        if c == i:
            imprime = [i]
            return imprime
        else:
            imprime = c
    return imprime
#Llamar a la funcion

resultado = revisar(palabra, letra)
#Imprimimos
print(resultado)
print(type(resultado))
```
