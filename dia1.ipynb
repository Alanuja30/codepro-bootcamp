#### Operaciones Aritmeticas  con variables


```python
producto1 = 'papas fritas'
costo_producto1 = 15000

producto2 = 'coca cola'
costo_producto2 = 14000

capital = 50000
dolar = 6700

# Operacion suma
costo_total = costo_producto1   +   costo_producto2
print (costo_total)

# Operacion resta
vuelto = capital - costo_total
print (vuelto)

# Realizamos la division entera
vuelto_dolar = vuelto//dolar
print(vuelto_dolar)
```

```python
#  variables
cantidad_de_manzanas = 100     
cantidad_de_naranjas = 300

# Operacion Suma
cantidad_de_frutas = cantidad_de_manzanas + cantidad_de_naranjas

#Imprimir Cantidad de Frutas
print('Cantidad de frutas: ', cantidad_de_frutas)

# Operacion Resta
frutas_vendidas = 150
frutas_restantes = cantidad_de_frutas - frutas_vendidas

#Imprimir Cantidad de Frutas Restantes
print('\nCantidad de frutas restantes: ', frutas_restantes)
```
