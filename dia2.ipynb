### Challenge 游뱤 - 20m
Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:

- Entre -10 y 18 grados, que imprima "Fresquete".
- Entre 19 y 30 grados, que imprima "Calor'i".
- Entre 31 y 45 grados, que imprima "Hakuuuuuu".
- Para todo lo dem치s, imprimir "칌amano mba'e!".

```python

# Pedimos al usuario que ingrese la temperatura e imprimimos
temperatura = int(input("Ingrese la temperatura en grados Celsius: "))
print(f"La Temperatura es de {temperatura} 췈C")

#Primera Condicion
if temperatura >= -10 and temperatura <= 18:
    print("Fresquete")
#Segunda Condicion
elif temperatura >=19 and temperatura <=30:
    print("Calor'i")
#Tercera Condicion
elif temperatura >=31 and temperatura <=45:
    print('Hakuuuuu')
#Ultima Condicion
else:
    print("칌amano mba'e!")
```

#### Bucle while y while True

Escribir un programa que almacene una contrase침a en una variable, y pregunte al usuario por la contrase침a hasta que introduzca la contrase침a correcta.

Cuando el usuario ingrese la contrasena correcta, imprimir un mensaje secreto.

```python
# Guardamos la contrase침a
contrasena = "penguin"
# Preguntamos por primera vez al usuario por la contrase침a
contrasena_usuario = input('Ingrese contrase침a: ')

#Mientras la contrasena ingresada sea distinta vuelve a preguntar al preguntar
while contrasena != contrasena_usuario:
    print('Acceso Denegado - vuelva intentar')
    contrasena_usuario = input('Ingrese contrase침a: ')
#Sino se cumple la condicion la contrase침a es correcta
else:
    print("Acceso Correcto -  Bienvenido!")
```

#### Bucles iterativos for

Challenge 游뱤 - 10m
Escribir un programa que itere en un rango de 10 veces.
Por cada iteracion multiplicar el iterador por dos e imprimir el resultado. 

```python

for i in range(10):
    resultado = i*2     #Multiplicamos el iterador por 2
    print(f'El iterador {i} por 2 da resultado: {resultado}')  #Imprimimos el resultado y vuelve a iniciar el bucle 
```

#### Challenge
Solicitar al usuario que ingrese una frase
Imprimir un mensaje cada vez que se encuentre una letra 'o'

```python
#Pedimos al usuario que ingrese la frase, guardamos e imprimimos
frase = input('Ingrese una frase: ')
print('La frase es: ', frase)

#Definimos una variable contador
contar = 0
# Recorremos la frase
for letra in frase:
    if letra == 'o':
        contar = contar +1
        print ('Se encontro la o en: ', frase)

print(f"La letra 'o' aparece {contar} veces en la frase")       3
```
