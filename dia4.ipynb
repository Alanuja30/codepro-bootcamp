- Atributos de instancia: `nombre` & `color`

```python
#Definimos la clase
class Animal:
# Atributos de clase
    especie = 'canino'
#Atributo de Instancias
    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color

perro1 = Animal('Holly','marron')
perro2 = Animal('Anya','negro con marron')

print(f'Me llamo {perro1.nombre} y soy un {Animal.especie} de color {perro1.color}')
print(f'Me llamo {perro2.nombre} y soy un {perro2.especie} de color {perro2.color}')
```

### Challenge 2ü§∫ - 15 min

Ejemplo: "Cuando hablo, digo X".
Ejemplo: "Cuando me muevo, hago X".

```python
class Mascota:
#Atributos de clase
    especie = 'canina'

    #Atributos de Instancias
    def __init__(self, nombre):
        self.nombre = nombre

    #Metodos
    def ladra(self):
        print(f'{self.nombre} esta ladrando')

    def caminar(self, pasos= 0):
        print(f'{self.nombre} camina {pasos} pasos')

perro1 = Mascota('Holly')
perro1.ladra()
perro1.caminar(10)
```

#### Challenge 4 ü§∫ - 20 min
Crear dos personajes de una historia que se representen a traves de clases. Ejemplo: Un hereo y un villano.
Cada personaje debe tener poderes o acciones diferentes. Y deben presentarse en una frase.

Ejemplo: Yo me llamo pepito y tengo super fuerza
Ejemplo: Yo me llamo juanito y puedo volar

```python

# Definimos la Clase
class Personaje:
    def __init__ (self, nombre, poder):
        self.nombre = nombre
        self.poder = poder

#Definimos las Clases a Heredar
#Clase Villano
class Villano(Personaje):
    def villano (self):
        print(f'Soy un {self.nombre} y mis poderes son {self.poder}')

#Clase Heroe
class Heroe(Personaje):
    def heroe (self):
        print(f'Soy un {self.nombre} y mis poderes son {self.poder}')

#Completamos datos
villano1 = Villano('Lex Luthor','Super Inteligecia')
heroe1 = Heroe('Superman', 'Super Fueza y Vuelo')

#Ejecutamos los Metodos
heroe1.heroe()
villano1.villano()
```
